Варіант 10
$$
  \left\{
  \begin{aligned}
     & –10.2 x_1 + 9.3 x_2 = 73 \\
     & 5.3 x_1 + 15.8 x_2 = 173 \\
     & 17.2 x_1 + 10.3 x_2 = 159 
  \end{aligned}
  \right.
$$

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
f0 = lambda x: (73 + 10.2 * x) / 9.3
f1 = lambda x: (173 - 5.3 * x) / 15.8
f2 = lambda x: (159 - 17.2 * x) / 10.3

x = np.linspace(0, 5, 400)
y1 = f0(x)
y2 = f1(x)
y3 = f2(x)
y_min = min(np.min(y1), np.min(y2), np.min(y3))
y_max = max(np.max(y1), np.max(y2), np.max(y3))

plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.plot(x, f0(x), color='red', linestyle='dashed', label='$-10.2x_1 + 9.3x_2 = 73$')
plt.grid(True)
plt.title('$–10.2 x_1 + 9.3 x_2 = 73$')
plt.ylim(y_min, y_max)

plt.subplot(1, 3, 2)
plt.plot(x, f1(x), color='green', linestyle='dashdot', label='$5.3x_1 + 15.8x_2 = 173$')
plt.grid(True)
plt.title('$5.3 x_1 + 15.8 x_2 = 173$')
plt.ylim(y_min, y_max)

plt.subplot(1, 3, 3)
plt.plot(x, f2(x), color='blue', linestyle='dotted', label='$17.2x_1 + 10.3x_2 = 159$')
plt.grid(True)
plt.title('$17.2 x_1 + 10.3 x_2 = 159$')
plt.ylim(y_min, y_max)

plt.tight_layout()
plt.show()
```

```python
x_intersection_1_2 = (73*15.8 - 9.3*173) / (15.8*(-10.2) - 9.3*5.3)
y_intersection_1_2 = f0(x_intersection_1_2)

x_intersection_2_3 = (173*10.3 - 15.8*159) / (10.3*5.3 - 15.8*17.2)
y_intersection_2_3 = f1(x_intersection_2_3)

x_intersection_1_3 = (159*9.3 - 10.3*73) / (9.3*17.2 + 10.2*10.3)
y_intersection_1_3 = f2(x_intersection_1_3)

plt.figure(figsize=(8, 6), dpi=100)

plt.plot(x, y1, color='red', linestyle='dashed', label='f[1]')
plt.plot(x, y2, color='green', linestyle='dashdot', label='f[2]')
plt.plot(x, y3, color='blue', linestyle='dotted', label='f[3]')

plt.xlabel('$x_1$')
plt.ylabel('$x_2$')
plt.title('All lines in one figure')
plt.legend()
plt.grid(True, color='orange', linestyle='--')
plt.text(0, 8, '$-10.2x_1 + 9.3x_2 = 73$', color='red', rotation=26)
plt.text(0, 10.7, '$5.3x_1 + 15.8x_2 = 173$', color='green', rotation=-8)
plt.text(0, 13.8, '$17.2x_1 + 10.3x_2 = 159$', color='blue', rotation=-37)

plt.fill_between(x, y2, y_max, color='gray', alpha=0.3)
plt.fill_between(x, y1, y_max, color='white', alpha=1.0)
plt.fill_between(x, y3, y_max, color='white', alpha=1.0)

A1 = np.array([[-10.2, 9.3], [5.3, 15.8]])
A2 = np.array([[-10.2, 9.3], [17.2, 10.3]])
A3 = np.array([[5.3, 15.8], [17.2, 10.3]])
B1 = np.array([73, 173])
B2 = np.array([73, 159])
B3 = np.array([173, 159])
intersections1 = np.linalg.solve(A1, B1)
intersections2 = np.linalg.solve(A2, B2)
intersections3 = np.linalg.solve(A3, B3)

plt.plot(intersections1[0], intersections1[1], 'ro')
plt.plot(intersections2[0], intersections2[1], 'ro')
plt.plot(intersections3[0], intersections3[1], 'ro')

offset1 = 1
offset2 = 3
offset3 = -2;

plt.text(intersections1[0], intersections1[1] + offset1, f'({intersections1[0]:.2f}, {intersections1[1]:.2f})', ha='right')
plt.text(intersections2[0], intersections2[1] + offset2, f'({intersections2[0]:.2f}, {intersections2[1]:.2f})', ha='right')
plt.text(intersections3[0], intersections3[1] + offset3, f'({intersections3[0]:.2f}, {intersections3[1]:.2f})', ha='right')

plt.plot([intersections1[0], intersections1[0]], [intersections1[1], intersections1[1] + offset1], color='black')
plt.plot([intersections2[0], intersections2[0]], [intersections2[1], intersections2[1] + offset2], color='black')
plt.plot([intersections3[0], intersections3[0]], [intersections3[1], intersections3[1] + offset3], color='black')

plt.show()
```
